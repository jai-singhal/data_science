```python
sentence = "My name is Jai Singhal."

grammar = ('''
    NP: {<NNP>+} # first name with surname
    ''')

chunkParser = nltk.RegexpParser(grammar)
tagged = nltk.pos_tag(nltk.word_tokenize(sentence))
tree = chunkParser.parse(tagged)

for subtree in tree.subtrees():
    print(subtree)
tree.draw()
```

```python
obituary = open("./obituaries.txt", "r")
sentence = obituary.read()
grammar = ('''
    NP: {<NNP>+} # first name with surname
    ''')

chunkParser = nltk.RegexpParser(grammar)
tagged = nltk.pos_tag(nltk.word_tokenize(sentence))
tree = chunkParser.parse(tagged)

for subtree in tree.subtrees():
    print(subtree)
# tree.draw()
```

```python
# Named entity recognition
obituary = "Michael Jackson was died on January 20,2002."
for sentence in nltk.sent_tokenize(obituary)[0:1]:
    print(sentence)
    words = nltk.word_tokenize(sentence)
    tagged = nltk.pos_tag(words)
    namedEntity = nltk.ne_chunk(tagged, binary=True)
    for subtree in namedEntity.subtrees():
        print(subtree)
    namedEntity.draw()
```

```python
# Lemetizing
from nltk import WordNetLemmatizer

lemmitizer = WordNetLemmatizer()

print(lemmitizer.lemmatize("cats"))
print(lemmitizer.lemmatize("dogs"))
print(lemmitizer.lemmatize("pythonista", pos = "a"))
print(lemmitizer.lemmatize("better", pos = "a"))
print(lemmitizer.lemmatize("better"))
print(lemmitizer.lemmatize("biggest", pos = 'v'))
print(lemmitizer.lemmatize("bigger"))
print(lemmitizer.lemmatize("worst", pos = 'a'))
print(lemmitizer.lemmatize("chatting"))
```
